```python
from transformers import MarianMTModel, MarianTokenizer
```

```python
# Configuración del modelo
model_name = "Helsinki-NLP/opus-mt-en-es"  # Modelo inglés-español
tokenizer = MarianTokenizer.from_pretrained(model_name)
model = MarianMTModel.from_pretrained(model_name)

# Traducción de ejemplo
text = ["The stars are shining brightly tonight.",
        "Dreams come alive under the night sky.",
        "A peaceful evening in the forest."]
translated = model.generate(**tokenizer(text, return_tensors="pt", padding=True, truncation=True))

# Mostrar traducciones
for i, t in enumerate(translated):
    print(f"Original: {text[i]}")
    print(f"Traducción: {tokenizer.decode(t, skip_special_tokens=True)}")
```
